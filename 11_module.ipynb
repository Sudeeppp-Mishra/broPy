# Module 

a file containing code you want to include in your program

USE
```python
import
```
to include a module (built-in or your own)

useful to break up a large program reusable separate files

````python
print(help("modules")) # modules we can use in pythons
````

````python
print(help("math")) # prints differnet functions and attributes in a module
````

````python
import math

print(math.pi)
````

````python
import math as m # m is a alias (like nickname for ease)

print(m.pi)
````

````python
from math import pi # importing specific thing

print(pi)
````

````python
from math import e

a, b, c, d = 1, 2, 3, 4

print(e**a)
print(e**b)
print(e**c)
print(e**d)

a, b, c, d, e = 1, 2, 3, 4, 5 # so we could have naming problem when we use `from ... import ...`

print(e**a)
print(e**b)
print(e**c)
print(e**d)
print(e**e)
````
